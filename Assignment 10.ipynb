**Q.1** How do you distinguish between shutil.copy() and shutil.copytree()?

**Answer :-**

  * shutil.copy() : Function will copy a single file and return destination  folder it may be a directory also.

  *  shutil.copytree() : Function will copy an entire  source directory , and return the destination directory.

**Q.5** Create a programme that searches a folder tree for files with a certain file extension (such as .pdf or .jpg). Copy these files from whatever location they are in to a new folder.


```python
import os, shutil

def SelectiveCopy(source, extensions, destFolder):
    folder = os.path.abspath(source)
    destFolder = os.path.abspath(destFolder)
    print('Looking in', source, 'for files with extensions of', ', '.join(extensions))
    for foldername, subfolders, filenames in os.walk(source):
        for filename in filenames:
            name, extension = os.path.splitext(filename)
            if extension in extensions:
                fileAbsPath = foldername + os.path.sep + filename
                print('Coping', fileAbsPath, 'to', destFolder)
                shutil.copy(fileAbsPath, destFolder)

extensions = ['.jpg', '.txt', '.xlsx']
source = "D:\python program\ineuron lecture"
destFolder = "D:\python program\ineuron lecture\myNewFolder"
SelectiveCopy(source, extensions, destFolder)
```
